```python
#The hashtag that I am searhing for is climate change and
#tweets going back 5 years.
search_words = "climate+change"
date_since = "2015-02-01"
```

```python
# Collecting the tweets (only tweets written in english). I have only extracted a 1000 tweets 
#because I cannot upload such a large file into GITHUB
tweets = tweepy.Cursor(api.search, q=search_words, lang="en", since=date_since).items(1000)
```

```python
#This extracts user screen name, location of where the user is/tweeted from, when the tweet was created 
#as well as the tweet itself
CC_Tweets = [[tweet.user.screen_name, tweet.user.location, tweet.text, tweet.created_at] for tweet in tweets]
```

```python
Climate_Change_DF = pd.DataFrame(data=CC_Tweets, 
                    columns=["User", "Location","Tweet","Date/Time of Tweet"])
```
